## F2 Carrer

Analysis of the F2 career of the F1 driver that come from it


```python
f2_data=pd.read_csv("DeVries Comparasion.csv")
```

```python
f2_data
```

```python
plt.subplots(figsize=(15, 10))
fig=sns.countplot(x='Driver',data=f2_data,order=f2_data['Driver'].value_counts().index)
plt.xlabel("Driver")
plt.ylabel("Number of Seasons")
plt.title("Number of Seasons in F2")
```

```python
plt.subplots(figsize=(15, 10))
fig=sns.barplot(x='Driver',y='Result',hue='Season',data=f2_data)
plt.xlabel("Driver")
plt.ylabel("Result per Season")
plt.title("Result per Season in F2")
plt.yticks(np.arange(0, 20, step=1)) 
```

```python
plt.subplots(figsize=(15, 10))
fig=sns.boxplot(x='Driver',y=f2_data['Result'],data=f2_data)
plt.xlabel("Driver")
plt.ylabel("Avg result per Season")
plt.title("Result per Season in F2")
plt.yticks(np.arange(0, 20, step=1)) 
```

```python
f2_data.groupby('Driver')['Result'].mean()
```

```python
plt.subplots(figsize=(15, 10))
fig=sns.boxplot(x='Driver',y=f2_data['Difference To Leader/Next Car'],data=f2_data)
plt.xlabel("Driver")
plt.ylabel("Gap to leader or 2nd place per Season")
plt.title("Average Gap to leader or 2nd place in F2")
```

```python
f2_data.groupby('Driver')['Difference To Leader/Next Car'].mean()
```

```python
plt.subplots(figsize=(15, 10))
fig=sns.boxplot(x='Driver',y=f2_data['Points'],data=f2_data)
plt.xlabel("Driver")
plt.ylabel("Points per Season")
plt.title("Avg points per season in F2")
```

```python
f2_data.groupby('Driver')['Points'].mean()
```

## Pace Vs Russell (Mercedes, FP1 Spain 2022)

```python
year, grand_prix, session = 2022, 'Spain', 'FP1'
```

```python
fp1_spain=ff1.get_session(year, grand_prix, session)
fp1_spain.load()
```

```python
devr=fp1_spain.laps.pick_driver('DEV')
russ=fp1_spain.laps.pick_driver('RUS')
```

```python
ff1.plotting.setup_mpl(mpl_timedelta_support=True)
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
ax.plot(devr['LapNumber'], devr['LapTime'], color='red')
ax.plot(russ['LapNumber'], russ['LapTime'], color='cyan')
ax.set_title("DeVries vs Russell")
ax.set_xlabel("Lap Number")
ax.set_ylabel("Lap Time")
plt.show()

#DEVRIES IN RED vs RUSSEL IN BLUE
```
